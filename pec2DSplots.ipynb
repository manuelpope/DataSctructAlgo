```python
import pandas as pd


with open('/content/output.txt') as f:
    lines = f.readlines()
```

```python
lines = [line.split('\t|') for line in lines  ]
```

```python
lines
```

```python
pa= [line[3].replace('\n','') for line in lines  ]
qs =[line[2].replace('\n','') for line in lines  ]
ms =[line[1].replace('\n','') for line in lines  ]
n =[line[0].replace('\n','') for line in lines  ]
pa= [line[3].replace('AS: ','') for line in lines  ]
qs =[line[2].replace('QS: ','') for line in lines  ]
ms =[line[1].replace('MS: ','') for line in lines  ]
n =[line[0].replace('SZ: ','') for line in lines  ]
```

```python
pa= [float(line) for line in pa  ]
qs =[float(line) for line in qs  ]
ms =[float(line) for line in ms  ]
n =[float(line)for line in n  ]


```

```python
d = {'n': n, 'ms':ms,'pa':pa,'qs':qs}
df = pd.DataFrame(data=d)
df.head(30)
```

```python
dfMean= df.groupby(['n']).mean()
dfMean['ms'] = dfMean['ms']/1000
dfMean['qs'] = dfMean['qs']/1000
dfMean['pa'] = dfMean['pa']/1000
dfMean
```

```python
import matplotlib.pyplot as plt
df.plot(x="n", y=["ms",'qs','pa'])

plt.show()
```
